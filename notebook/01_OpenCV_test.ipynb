# 画像をOpenCVで読み込み、表示できるかテストする

```python
#coding:utf-8
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# matplotlibをJupyterでインライン表示するための宣言
%matplotlib inline
```

## 設定

```python
NAMECARD_FOLDER = "../data/01_namecard/"
```

## OpenCVでGrayscale表示

```python
def plot_cv_grayscale_image(img_path): 
    img_src = cv2.imread(img_path, 1)
    img_obj_np = cv2.cvtColor(img_src, cv2.COLOR_BGR2GRAY)    # グレースケールに変換
    fig, ax = plt.subplots(figsize=(14, 11))
    plt.imshow(img_obj_np)  #貼り付け
    plt.gray() # Grayscale表示に切り替え
    plt.show()   #表示
```

```python
test_path = NAMECARD_FOLDER + "001_hugkun_yasubei.jpg"
plot_cv_grayscale_image(test_path)
```

## OpenCVでColor表示

```python
def plot_cv_color_image(img_path): 
    img_src = cv2.imread(img_path, 1)
    fig, ax = plt.subplots(figsize=(14, 11))
    plt.imshow(img_src)
    plt.show()   #表示
```

```python
plot_cv_color_image(test_path)
```

## PILでカラー表示

```python
def plot_image(img_path):
    im = Image.open(img_path)    #画像の読み込み
    im_list = np.asarray(im)           #画像をarrayに変換
    fig, ax = plt.subplots(figsize=(14, 11))
    plt.imshow(im_list)                   #貼り付け
    plt.show()                                   #表示
```

```python
plot_image(NAMECARD_FOLDER + "001_hugkun_yasubei.jpg")
```
